```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
data= yf.download("AAPL",start='2023-12-12',end='2024-12-12')
closingdata= data["Close"]
plt.plot(closingdata.index, closingdata, color='blue', label='Closing Price')
plt.title(f'{"AAPL"} Closing Price - Past 1 Year')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
#the general trend of the stock is upwards, there was a significant dip in the first part of the year but 
#once it started rising it didnt stop and ended on a high trend
```
