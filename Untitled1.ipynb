```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python



tickers = ['AAPL', 'MSFT', 'GOOGL'] 
data = {}

for ticker in tickers:
    data[ticker] = yf.download(ticker, start='2024-06-12', end='2024-12-12')['Close']



investment_per_stock = 3333.33


shares = {ticker: investment_per_stock / data[ticker].iloc[0] for ticker in tickers}


num_data = len(data[tickers[0]])
portfolio_returns = np.zeros(num_data)


for i in range(1, num_data):
    
    daily_returns = np.array([(data[ticker].iloc[i] - data[ticker].iloc[i-1]) / data[ticker].iloc[i-1] for ticker in tickers])
    
    
    portfolio_returns[i] = np.sum([shares[ticker] * daily_returns[j] for j, ticker in enumerate(tickers)])


plt.figure(figsize=(10, 6))
plt.plot(range(num_data), portfolio_returns, label='Portfolio Returns', color='b')
plt.title('Portfolio Returns Over 6-Month Period')
plt.xlabel('Days')
plt.ylabel('Portfolio Returns ($)')
plt.legend()
plt.grid(True)
plt.show()

    

             
             
                 
    
```
